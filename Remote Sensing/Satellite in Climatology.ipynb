## Building algorithm

### 20th percentile

```python
# Create empthy lists

date=[]
q2=[]
val=[]
lat=[]
lon=[]

###############################################################
# Calculate the quartiles of each pixel for whole timesteps.
# Then, precipitation value of each pixel is calculated for the relevant month.
# Put them all into separate lists if precipitation is smaller than threshold value.

for d in np.arange(0,ds_new.time.shape[0]): # Since "time" is not float, I must use "shape", instead of "len"
    for x in np.arange(0, len(ds_new.lon)):           
        for y in np.arange(0, len(ds_new.lat)):
        
            q_2=ds_new["precip"].isel(lon=x, lat=y).quantile(0.2).values
            v=ds_new["precip"].isel(lon=x, lat=y, time=d).values
            q=float(q)
            v=float(v)
            
            if v<q_2:
                
                date.append(str(ds_new.time[d]))
                lon.append(str(ds_new.lon[x]))
                lat.append(str(ds_new.lat[y]))
                q2.append(q_2)
                val.append(v)
                
            else:
                break
                
```

### Creating Dataframe

```python
d = {'Date': date, 'Montly Precipitation': val,
     "20th Percentile":q2,
     "lat":lat, "lon":lon}

df = pd.DataFrame(data=d)
df
```

```python
# Let's extract only date from the "Date" column

for i in np.arange(0, len(df.Date)):
    
    df.Date[i] = df.Date[i][36:43]
    df.lat[i] = df.lat[i][34:38]
    df.lon[i] = df.lon[i][34:38]

df
```

```python
df.index = pd.to_datetime(df.Date, format="%Y-%m")
df.drop(["Date","Unnamed: 0"], axis=1, inplace=True)
df['Year'] = df.index.year 
df['Month'] = df.index.month
df.reset_index(inplace=True)
df.to_csv(path+"europe.csv")
df
```

### Converting to NetCDF format

```python
df=pd.read_csv(path+"europe.csv", sep=",", usecols=[2,3,4,5,6])
df
```

```python
df["coord"]=""

for i in np.arange(0,len(df)):
    
    df["coord"][i]=str(df["lat"][i])+"-"+str(df["lon"][i])

df
```

```python
df19=df[(df.Year>=1990)&(df.Year<2000)].coord.value_counts().reset_index()
df19.to_csv(path+"df19.csv")
df20=df[(df.Year>=2000)&(df.Year<2010)].coord.value_counts().reset_index()
df20.to_csv(path+"df20.csv")
df21=df[(df.Year>=2010)&(df.Year<2020)].coord.value_counts().reset_index()
df21.to_csv(path+"df21.csv")
```

```python
dfs=[19,20,21]

for i in dfs:

    df=pd.read_csv(path+"df"+str(i)+".csv")
    df["lat"]=""
    df["lon"]=""

    df["lat"] = df["index"].str.split("-").str[0]
    df["lon"] = df["index"].str.split("-").str[1]

    df.drop("index", axis=1, inplace=True)
    df = df.rename(columns={'coord': 'freq'})

    df["lat"]=df["lat"].astype(float)
    df["lon"]=df["lon"].astype(float)
    
    dx = df.set_index(['lat','lon']).to_xarray()
    dx.to_netcdf(path+"dx"+str(i)+".nc")
    
```

Let me explain what I've performed here:

- merged latitude and longitude in the "coord" column
- sort coord column by year and then by count
- divide coord column into lat and lon again
- afterthat, converte the dataframes to NetCDF files
- finally, I have datasets and NetCDF files including pixel coordinates and their frequency
